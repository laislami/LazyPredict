```python
#importando as bibliotecas
import pandas as pd
import numpy as np
import lazypredict
from lazypredict.Supervised import Lazy.Regressor
from sklearn.model_selection import train_test_split
```

```python
#carregando e abrindo o dataset
dataset = pd.read_csv(" ")
dataset.head()
```

```python
#selecionar colunas e variável alvo
x = np.array(dataset.drop(['Sales'], 1))
y = np.array(dataset['Sales'])
```

```python
#dividindo a base de dados entre treinamento e teste
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 13))
```

```python
#o módulo lazypredict ajusta os dados a todos os modelos de regressão disponíveis no scikit-learn
reg = LazyRegressor(verbose=0, ignore_warnings=true, custom_metric = none)
models,predictions = reg.fit(x_train, x_test, y_train, y_test)
display(models)
```
